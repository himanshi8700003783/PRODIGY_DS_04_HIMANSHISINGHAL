```python
# Sentiment distribution 
sentiment_counts = df['Sentiments'].value_counts()
print(sentiment_counts)
plt.figure(figsize=(8, 5))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Brand Sentiment count % of Microsoft
brand_data=df[df['Entity'].str.contains('Microsoft',case=False)]
brand_sentiment_counts=brand_data['Sentiments'].value_counts()
brand_sentiment_counts
```

```python
plt.figure(figsize=(6,6))
plt.title("Brand Sentiment count % of Microsoft")
plt.pie(brand_sentiment_counts,labels=brand_sentiment_counts.index,autopct='%1.2f%%',startangle=140,colors=['cyan','#2ecc71','#e74c3c','yellow'],explode=[0, 0.1,0,0])
plt.show()
print("we can clearly say 35.42 % have neutral sentiment for microft, 24.87% positive while 32.47 negative sentiments")
```

```python
# Brand Sentiment count of Amazon
brand_data=df[df['Entity'].str.contains('Amazon',case=False)]
brand_sentiment_counts=brand_data['Sentiments'].value_counts()
brand_sentiment_counts
```

```python
plt.figure(figsize=(6,6))
plt.title("Brand Sentiment count of Amazon")
brand_sentiment_counts.plot(kind="bar" ,color=['red','cyan','yellow','pink'])
print("thus Amazon has more neutral sentiments than microsoft but microsoft has more positive sentiments")
```

```python
# Heatmap of Sentiment by Entity
entity_sentiment = pd.crosstab(df['Entity'], df['Sentiments'])
plt.figure(figsize=(10, 6))
sns.heatmap(entity_sentiment, annot=True, cmap='coolwarm', fmt='d')
plt.title('Sentiment by Entity')
plt.show()
```

```python
# Cat Plot for various entities displaying sentiments of each entity
plt.figure(figsize=(60, 20))
sns.catplot(x="Entity",hue="Sentiments", kind="count",height=10,aspect=4, data=df)
```

```python
def get_unique_values(input_string):  
  
    elements = input_string.split()   
    unique_values = set(elements)  
     
    return list(unique_values)  
```

```python
# Word Cloud of Positive and Negative Sentiments
from wordcloud import WordCloud
positive_text = ' '.join(df.Entity[df['Sentiments'] == 'Positive'])  
input_string = positive_text
unique_values = get_unique_values(input_string) 
u= ' '.join(map(str,unique_values))
u
positive_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(u)
plt.figure(figsize=(8, 5))
plt.imshow(positive_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Positive Sentiment Word Cloud')
plt.show()
```

```python
negative_text = ' '.join(df.Entity[df['Sentiments'] == 'Negative'])  
input_string = negative_text
unique_values = get_unique_values(input_string) 
u= ' '.join(map(str,unique_values))
u
negative_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(u)
plt.figure(figsize=(8, 5))
plt.imshow(negative_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Negative Sentiment Word Cloud')
plt.show()
```
